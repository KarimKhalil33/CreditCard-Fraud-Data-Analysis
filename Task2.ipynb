## 2.5 Answer your research questions with visualizations

Remember to refer to the [`seaborn` gallery](https://seaborn.pydata.org/examples/index.html) for ideas on different visualizations.

````python
df = pd.read_csv('https://github.com/firasm/bits/raw/master/pokemon.csv')

# Calculate the mean speed for each type
mean_speed = df.groupby('Type 1')['Speed'].mean().sort_values(ascending=False)

# Create a bar plot of the mean speed for each type
plt.figure(figsize=(12,6))
plt.bar(mean_speed.index, mean_speed.values, color='purple')

# Set plot and axis titles
plt.title('Average Speed of Each Pokemon Type')
plt.xlabel('Pokemon Type')
plt.ylabel('Speed')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
````

````python
df = pd.read_csv('https://github.com/firasm/bits/raw/master/pokemon.csv')

mean_defense = df.groupby('Type 1')['Defense'].mean().sort_values(ascending=False)

# Create a bar plot of the mean defense for each type
plt.figure(figsize=(12,6))
plt.bar(mean_defense.index, mean_defense.values, color='green')

# Set plot and axis titles
plt.title('Average Defense of Each Pokemon Type')
plt.xlabel('Pokemon Type')
plt.ylabel('Defense')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
````
